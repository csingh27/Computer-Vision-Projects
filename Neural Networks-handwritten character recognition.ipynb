# Character recognition using neural networks

## Background

### Input ### 
Given Image
### Process ### 
Each pixel of an image is fed as an input to the neural network with a single hidden layer
### Output ### 
Number corresponding to the image
### Steps ###
-

```python
import tensorflow as tf
import matplotlib.pyplot as plt
mnist=tf.keras.datasets.mnist
(x_train, y_train), (x_test, y_test)=mnist.load_data()
plt.figure()
plt.imshow(x_train[0])
plt.colorbar() #Colour Range from 0 to 255
plt.grid(False)
plt.show()
x_train, x_test = x_train / 255.0, x_test / 255.0 #Normalizing the data set 
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Flatten(input_shape=(28, 28)), #28*28=784 input neurons connected to all 10 neurons
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(10, activation='softmax') # a single hidden layer with 10 neurons
])
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
history=model.fit(x_train,y_train, epochs=25, validation_data=(x_test, y_test))
acc_train=history.history["accuracy"]
acc_test=history.history["val_accuracy"]
plt.show()
epochs=range(1,len(acc_train)+1)
plt.plot(epochs,acc_train,'r',label='Training accuracy')
plt.plot(epochs,acc_test,'b',label='Validation accyracy')
plt.title('Training and Validation accuracy')
plt.xlabel('epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```
